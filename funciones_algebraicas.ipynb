# Funciones algebraicas

## Función lineal

Tiene la forma de
$$f(x) = mx + b$$
donde $m$ y $b$ $\in R$

$m$ puede ser calculada por:
$$m=\frac {y_{2} - y_{1}}{x_{2} - x_{1}}$$

y $b$ es el punto de corte con el eje $y$. Su dominio es $Dom_{f} = (-\infty, \infty)$. Su imagen es $Im_{f} = (-\infty, \infty)$

```python
N = 100 # numero de valores a graficar


def f(x):
    return m*x + b

m = 1
b = 5

x = np.linspace(-10, 10, N)
y = f(x)
```

Graficamos la línea. <br>
Cuando $m$ es positiva ($+$) la pendiente incrementa

```python
plt.plot(x,y)
plt.grid()
```

Cuando $m$ es negativa ($-$) la pendiente decrece.

```python
m = -1
y = f(x)

plt.plot(x,y)
plt.grid()
```


Cuando $m = 0$ no hay pendiente, obtenemos una línea recta horizontal dado por el punto de corte $b$.

```python
m = 0
y = f(x)

plt.plot(x,y)
plt.grid()
```

$b$ es el punto de corte de la recta, y es donde la le recta toca el punto 0  del eje $x$

```python
m = 1
b= 3
y = f(x)

plt.plot(x,y)
plt.plot(0, b, 'ro')
plt.grid()
```
