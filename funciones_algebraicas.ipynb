# Funciones algebraicas

## Función lineal

Tiene la forma de
$$f(x) = mx + b$$
donde $m$ y $b$ $\in R$

$m$ puede ser calculada por:
$$m=\frac {y_{2} - y_{1}}{x_{2} - x_{1}}$$

y $b$ es el punto de corte con el eje $y$. Su dominio es $Dom_{f} = (-\infty, \infty)$. Su imagen es $Im_{f} = (-\infty, \infty)$

```python
N = 100 # numero de valores a graficar


def f(x):
    return m*x + b

m = 1
b = 5

x = np.linspace(-10, 10, N)
y = f(x)
```

Graficamos la línea. <br>
Cuando $m$ es positiva ($+$) la pendiente incrementa

```python
plt.plot(x,y)
plt.grid()
```

Cuando $m$ es negativa ($-$) la pendiente decrece.

```python
m = -1
y = f(x)

plt.plot(x,y)
plt.grid()
```


Cuando $m = 0$ no hay pendiente, obtenemos una línea recta horizontal dado por el punto de corte $b$.

```python
m = 0
y = f(x)

plt.plot(x,y)
plt.grid()
```

$b$ es el punto de corte de la recta, y es donde la le recta toca el punto 0  del eje $x$

```python
m = 1
b= 3
y = f(x)

plt.plot(x,y)
plt.plot(0, b, 'ro')
plt.grid()
```

## Función polinómica

Tiene la forma
$$P(x) = a_{n}x^n + a_{n-1}x^n-1 + ... + a_{2}x^2 + a_{1}x + a_{1}$$
A una función que tiene esta forma se le llama _polnomio_ de grado $n$. A los elemen tos $a$ se les llama coeficientes donde $a$ $\in R$ <br>
**Por ejemplo**:
$$P(x) = 2x^7 - x^4 + 3x^2 + 4$$
A esta función se le conoce como _polinomio de grado 7_.

```python
def f(x):
    return (2*x**7) - (x**4) + (3*x**2) + 4

y = f(x)
plt.plot(x,y)
plt.grid()
```

Cuando el grado de un polinomio es un numero impar la forma de la grafica es como la anterior (en forma de "serviente"). <br>
Sin embargo si el grado es un numero par, la grafica tiene la forma de una parabola.

```python
def f(x):
    return (2*x**6) - (x**4) + (3*x**2) + 4

y = f(x)
plt.plot(x,y)
plt.grid()
```

## Función potencia

A la función potencia también se le llama monomio, ya que tiene solo un termino en la expresión, tienen la siguiente forma:
$$ f(x) = x^a, a \in R$$

**Por ejemplo:**
$$ f(x) = x^2$$

El dominio de la función anterior es $Dom_{f} = (-\infty, \infty)$. Su imagen es $Im_{f} = [0, \infty)$

```python
def f(x):
    return x**2

y = f(x)

plt.plot(x, y)
plt.grid()
```

Al igual que los polinomios, cuando el exponente es un numero pas genere una parabola y cuando es impar crea una línea en forma de "serpiente"

```python
def f(x):
    return x**3

y = f(x)

plt.plot(x, y)
plt.grid()
```
