# Medidas de dispersión

En este notebook se repasan las medidas de dispersión: _varianza_ y la _desviación estandas_ con las librerias `pandas` y `numpy` utilizando el Dataset `iris`.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_iris = sns.load_dataset("iris")
df_iris.head()
```

### Medidas de dispersión para la variable `sepal length`:

```python
mean = df_iris["sepal_length"].mean()
var = df_iris["sepal_length"].var()
std = df_iris["sepal_length"].std()

print(f"Media: {mean}")
print(f"Varianza: {var}")
print(f"Desviación estándar: {std}")
```

En la población total, la varianza es de `0.685...` y la desviación estandas es de `0.828...`, mientras que el valro de la media es de `5.843`.

Las medidas de dispersión que obtuvimos de la variable `sepal length` a simple vista son pequeñas, sin embargo al observar el valor de la media vemos que también es un número pequeño, por lo tanto la dispersión es bastante alta.

Lo comprobaremos de forma grafica:

```python
sns.displot(data=df_iris, x="sepal_length", kind='kde')
plt.show()
```

En el diagrama de densidad anterior, se observa una distribución normal con una curva amplica, lo cual indica que su disperción es amplia como lo vimos en los cálculos anteriores.

### Medidas de dispersión muestral para la variable `sepal length`:

```python
iris_sample = df_iris.sample(frac=0.4)
s_mean = iris_sample["sepal_length"].mean()
s_var = iris_sample["sepal_length"].var()
s_std = iris_sample["sepal_length"].std()

print(f"Media muestral: {s_mean}")
print(f"Varianza  muestral: {s_var}")
print(f"Desviación estándar  muestral: {s_std}")
```

```python
sns.displot(data=iris_sample, x="sepal_length", kind='kde')
plt.show()
```
