# Combinación lineal

En algebra lineal, una combinación lineal de un conjunto de vectores, es un vector que se puede formar sumando las multiplicaciones escalaras de estos vectores.

Se expresa como:
$$
w = c_1v_1 + c_2v_2 + ... + c_nv_n 
$$
Donde:
- $c_1$ ... $c_n$: Son los escalares o coeficientes.
- $v_1$ ... $v_n$: Son los vectores.

Ejemplo:

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
c1, c2 = 2, 1
v1 = np.array([2, 5])
v2 = np.array([3, 2])

w = c1*v1 + c2*v2
print(f"Combinación lineal resultante: {w}")
```
