# Ejemplos de sistemas de ecuaciones: sin solucion, con una solución y con soluciones infinitas

Podemos encontrarnos con distintos sistemas de ecuaciones lineales, sin embargo no todos tiene solución o bien una única solucinón.

```python
# %matplotlib notebook
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

## Sin solución

Los sistemas de ecuaciones lineales que no tienen solución son aquellos done tenemos más ecuaciones que variables. 

Ejemplo:
$$
\begin{cases}      
  y = 3x + 5 \\      
  y = -x + 3 \\
  y = 2x + 1
\end{cases}
$$
Tenemos 3 ecuaciones y solo 2 variables.

Otra caracteristica de los sistemas de ecuaciones sin solución, es a la hora de graficarlas no hay un punto de intersección entre todas las ecuaciones que conforman el sistema.

📌 A este tipo de sistemas se le conoce como: **Sistema sobre detemrinado**.


### Grafica de un sistema de ecaciones lineales sin solución

```python
# Definimos las variables y las ecuaciones
x = np.arange(-6, 6)
y_1 = 3 * x + 5
y_2 = -1 * x + 3
y_3 = 2 * x + 1
```

```python
# Creamos las graficas
plt.figure()

plt.plot(x, y_1)
plt.plot(x, y_2)
plt.plot(x, y_3)

plt.xlim(-8, 8)
plt.ylim(-8, 8)

plt.grid(True)
plt.show()
```

## Solución única

Un sistema de ecuaciones lineales con una única solución es cuando hay exactamente un conjunto de valores para las variables que satisfacen a todas las ecuaciones simultaneamente.

Ejemplo:
$$
\begin{cases}   
  y = -x + 3 \\
  y = 2x + 1
\end{cases}
$$

Al graficar las ecuaciones sus respectivas rectas de intersectan en un único punto; lo que significa que existe solo una solución.

📌 A este sistema se le llama: **consistente y determinado**

### Grafica de un sistema de ecaciones lineales con una solución única

```python
x = np.arange(-6, 6)
y_2 = -1 * x + 3
y_3 = 2 * x + 1
```

```python
# Creamos las graficas
plt.figure()
plt.plot(x, y_2)
plt.plot(x, y_3)

plt.xlim(-8, 8)
plt.ylim(-8, 8)

plt.grid(True)
plt.show()
```

## Infinitas soluciones

Tenemos un sistema de ecuaciones con infinitas soluciones cuando may más de un conjunto (infinitos) de valores que satisfacen a las ecuaciones simultaneamente; o bien solo tenemos una ecuación en el sistema.

Ejemplo con una única ecuación:
$$
\begin{cases}
  y = 2x + 1
\end{cases}
$$

Ejemplo con dos ecuaciones:
$$
\begin{cases}   
  y = -x + 2 \\
  y = \frac {2x + 4} {2}
\end{cases}
$$

Cuando graficamos un sistema de ecuaciones lineales con infinitas soluciones en el plano podemos observar una sola recta.

📌 A este sistema se le llama: **consistente e indeterminado**

### Una única ecuación

```python
x = np.arange(-6, 6)
y_3 = 2 * x + 1
```

```python
# Creamos las graficas
plt.figure()
plt.plot(x, y_3)

plt.xlim(-8, 8)
plt.ylim(-8, 8)

plt.grid(True)
plt.title("Una ecuación")
plt.show()
```

```python
x = np.arange(-6, 6)
y_3 = -1 * x + 2
y_4 = (-2*x+4)/2
```

```python
# Creamos las graficas
plt.figure()
plt.plot(x, y_3)
plt.plot(x, y_4)

plt.xlim(-8, 8)
plt.ylim(-8, 8)

plt.grid(True)
plt.title("Dos ecuaciones")
plt.show()
```
