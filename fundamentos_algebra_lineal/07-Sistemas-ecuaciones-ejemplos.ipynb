## Sin soluci贸n

Los sistemas de ecuaciones lineales que no tienen soluci贸n son aquellos done tenemos m谩s ecuaciones que variables. 

Ejemplo:
$$
\begin{cases}      
  y = 3x + 5 \\

\end{cases}
$$
Tenemos 3 ecuaciones y solo 2 variables.

Otra caracteristica de los sistemas de ecuaciones sin soluci贸n, es a la hora de graficarlas no hay un punto de intersecci贸n entre todas las ecuaciones que conforman el sistema.

 A este tipo de sistemas se le conoce como: **Sistema sobre detemrinado**.


### Grafica de un sistema de ecaciones lineales sin soluci贸n

```python
# Definimos las variables y las ecuaciones
x = np.arange(-6, 6)
y_1 = 3 * x + 5
y_2 = -1 * x + 3
y_3 = 2 * x + 1
```

```python
# Creamos las graficas
plt.figure()

plt.plot(x, y_1)
plt.plot(x, y_2)
plt.plot(x, y_3)

plt.xlim(-8, 8)
plt.ylim(-8, 8)

plt.grid(True)
plt.show()
```

Un sistema de ecuaciones lineales con una 煤nica soluci贸n es cuando hay exactamente un conjunto de valores para las variables que satisfacen a todas las ecuaciones simultaneamente.

Al graficar las ecuaciones sus respectivas rectas de intersectan en un 煤nico punto; lo que significa que existe solo una soluci贸n.

 A este sistema se le llama: **consistente y determinado**

### Grafica de un sistema de ecaciones lineales con una soluci贸n 煤nica
