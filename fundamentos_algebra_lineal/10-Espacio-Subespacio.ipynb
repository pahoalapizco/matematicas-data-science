# Espacio vectorial y Subespacio vectorial

Un espacio vectorial es una estructura matemática que se comone de un conjunto de vectores y dos operaciones: adición de vectores y multiplicación por un escalar. <br>
¿Qué compone a un espacio vectorial?
- Vecrtores: Elementos que encontramos en el espacio vectorial.
- Escalares: Números que se utilizan para escalar a los vectores, pertenecen a los reales $\Reals$ o a los complejos $\Complex$

**Los espacios vectoriales deben cumplir con ciertas propiedades:**
- **Cierre bajo la adición**: La suma de dos vectores en el espacio es también un vector dentro del mismo espacio.
- **Cierre bajo la multiplicación por un escalar**: El producto de un vector en el espacio por un escalar es también un vector en el mismo espacio.
- **Otras propiedades**: Asociatividad, conmutatividad de la adición, existencia de un vector cero, etc.

📌 A un espacio vectorial también se le conoce como "espacio lineal".

## Subespacio vectorial

Es un subconjunto de un espacio vectorial, este a su vez es en sí mismo un espacio vectorial bajo las mismas operaciones: adición de vectores y multiplicación por escalares.

Al hablar de espacios y subespacios vectoriales, nos contramos con ${\Reals}^1$, ${\Reals}^2$ y ${\Reals}^3$, los cuales representan espacios vectoriales en dimensiones 1, 2 y 3 sobre el campo de los números reales (${\Reals}$).

- ${\Reals}^1$: Conjunto de todos los números reales, de dimensión 1, cada vector es solo un número real $x \in \Reals$.

- ${\Reals}^2$: Conjunto de todos los pares ordenados de números reales, $\begin {pmatrix} x\\ y \end{pmatrix}$, donde $x,y \in \Reals $, con dimensión 2.

- ${\Reals}^3$: Conjunto de todas las ternas ordenadas de los números reales, $\begin {pmatrix} x\\ y \\ z \end{pmatrix}$, donde $x,y,z \in \Reals $

Ejemplos sencillos de un espacio vectorial y subespacios:

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

Ejemplo 1:

```python
# Vectores
v1 = np.array([1, 1])
v2 = np.array([-1, -1])

# Grafica 
for a in range(-10, 10):
  for b in range(-10, 10):
    plt.scatter(
      x=(a*v1[0] + b*v2[0]),
      y=(a*v1[1] + b*v2[1]),
      marker=".",
      color="orange"
    )

plt.axvline(x=0, color="gray")
plt.axhline(y=0, color="gray")
plt.show()
```

La combinación lineal de los vectores $[1,1]$ y $[-1,-1]$ da como resultado una recta, la cual es el espacio en el que "viven" los vectores, pues combinación lineal de dos elementos, solo pueden devolver un elemento en el mismo lugar(_espacio_) donde ellos están viviendo.

Ejemplo 2:

```python
# Vectores
v1 = np.array([1, 0])
v2 = np.array([2, -3])

# Grafica 
for a in range(-10, 10):
  for b in range(-10, 10):
    plt.scatter(
      x=(a*v1[0] + b*v2[0]),
      y=(a*v1[1] + b*v2[1]),
      marker=".",
      color="orange"
    )

plt.axvline(x=0, color="gray")
plt.axhline(y=0, color="gray")
plt.show()
```

Ejemplo 3:

```python
# Vectores
v1 = np.array([1, 0, 0])
v2 = np.array([2, -3, 0])

# # Grafica 
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

for a in range(-10, 10):
  for b in range(-10, 10):
    ax.scatter(
      v1[0]*a + v2[0]*b,
      v1[1]*a + v2[1]*b,
      v1[2]*a + v2[2]*b,
      marker=".",
      color="orange"
    )

plt.show()
```
