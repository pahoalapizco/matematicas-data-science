```python
import numpy as np
import matplotlib.pyplot as plt
```

# Características de las funciones

Podemos encontrar varios tipos de funciones, una de ellas son las funciones reales, se llaman de esta manera ya que tanto su dominio (valores de entrada) como su rango (valores de salida) pertenecen al conjunto de los números reales.

A continuación se explican de manera breve algunas de las principales características de las funciones reales.

## Función par

Una función es par si cumple la siguiente relación a lo largo de su dominio:
$$ f(-x) = f(x)$$
Lo anterior significa que la función es simétrica al eje vertical $y$, y por lo tanto, su gráfica es simétrica respecto al origen.

Un ejemplo de una función par es $f(x) = x^2$, por que $f(-x) = (-x)^2 = x^2 = f(x)$

**Comprobación**

```python
N = 100

def f(x):
    return x ** 2

x = np.linspace(-15, 15, num=N)
y = f(x)
y_cmp = f(-x)

fig, ax = plt.subplots(1, 2)
ax[0].plot(x, y)
ax[0].grid()
ax[0].axhline(color='red')
ax[0].axvline(color='red')
ax[0].set_title('f(x)')

ax[1].plot(x, y_cmp)
ax[1].grid()
ax[1].axhline(color='red')
ax[1].axvline(color='red')
ax[1].set_title('f(-x)')

plt.show()
```

## Función impar

Una función es par si cumple la siguiente relación a lo largo de su dominio:
$$ f(-x) = -f(x)$$
Lo que significa que la función es simétrica respecto al origen.

Un ejemplo de una función impar es $f(x) = x^3$ por que $f(-x) = -(-x)^3 = -f(x) $

**Comprobación**

```python
N = 100

def f(x):
    return x ** 3

x = np.linspace(-15, 15, num=N)
y = f(-x)
y_cmp = -f(x)

fig, ax = plt.subplots(1, 2)
ax[0].plot(x, y)
ax[0].grid()
ax[0].axhline(color='red')
ax[0].axvline(color='red')
ax[0].set_title('f(-x)')

ax[1].plot(x, y_cmp)
ax[1].grid()
ax[1].axhline(color='red')
ax[1].axvline(color='red')
ax[1].set_title('-f(x)')

plt.show()
```

## Función acotada

Para este tipo de función podemos encontrar 3 variantes:
- Acotada superiormente
- Acotada inferiormente
- Acotada 
En este caso nos enfocaremos en la última.

Una función es acotada si su rango se encuentra entre dos número reales $-m$ y $m$ tal que
$$ -m <=f(x) <= m $$

En otras palabras, la función no crece indefinidamente.
Un ejemplo de este tipo de funciones son `seno` y `coseno` que estan acotadas por el intervalo $[-1, 1]$.

**Comprobación**

```python
N = 100

x = np.linspace(-10, 10, num=N)

y_s = np.sin(x)
y_c = np.cos(x)

fig, ax = plt.subplots(1, 2)

#Seno
ax[0].plot(x, y_s)
ax[0].grid()
ax[0].set_title('sin(x)')
ax[0].axhline(y=1, color='red')
ax[0].axhline(y=-1, color='red')

# Coseno
ax[1].plot(x, y_c)
ax[1].grid()
ax[1].set_title('cos(x)')
ax[1].axhline(y=1, color='red')
ax[1].axhline(y=-1, color='red')

plt.show()
fig.tight_layout()
```

## Funciones monótonas

Una función es monótona si siempre crece o siempre decrese a lo largo de su dominio.
Que algo sea monótono significa que no tiene variaciones, entonces las funciones monótonas son aquellas que dentro de un intervalo $I$, perteneciente a los números reales, cumple alguna de las siguientes propiedades:

### 1. Estrictamente creciente

$$
Si \space para \space todo \space x_1, x_2 \in I : x_1 < x_2 \Longleftrightarrow f(x_1) < f(x_2)
$$

Se lee como:
> Si para todo $x_1$ y $x_2$ que pertenecen al intervalor $I$, tal que $x_1$ sea menor a $x_2$ si y solo si la función de $x_1$ es menor a la función de $x_2$

Siempre aumenta en su conjunto de definición.

Un ejemplo es $f(x) = x^2$, es estrictamente creciente en su dominio de números reales positivos.

**Comprobación**

```python
N = 10

def f(x):
    return x ** 2

def verify(arr):
    comp = ""
    for idx, val in enumerate(arr):
        if idx < len(arr) - 1:
            comp += f'{arr[idx]} < {arr[idx+1]} ({arr[idx] < arr[idx+1]})  \n'
    return comp

x = np.linspace(1, 5, N)
y = f(x)

print('x: \n', verify(x))
print('f(x): \n',verify(y))
```

```python
fig, ax = plt.subplots(1, 2)

ax[0].plot(x, 'ro-')
ax[0].grid()
ax[0].set_title('x')

ax[1].plot(y, 'go-')
ax[1].grid()
ax[1].set_title('$f(x) = x^2$')

plt.show()
```

### 2. Estrictamente decreciente

$$
Si \space para \space todo \space x_1, x_2 \in I : x_1 < x_2 \Longleftrightarrow f(x_1) > f(x_2)
$$

Se lee como:
> Si para todo $x_1$ y $x_2$ que pertenecen al intervalor $I$, tal que $x_1$ sea menor a $x_2$ si y solo si la función de $x_1$ es mayor a la función de $x_2$

Siempre disminuye en su conjunto de definición. 

Un ejemplo es $f(x) = -x^3$, es estrictamente decreciente en su dominio de números reales negativos.

**Comprobación**

```python
N = 10

def f(x):
    return -x ** 3

x = np.linspace(-5, -1, num=N)
y = f(x)

fig, ax = plt.subplots(1, 2)
ax[0].plot(x, 'ro-')
ax[0].grid()
ax[0].set_title('x')
ax[0].axhline(color='blue')
ax[0].axvline(color='blue')

ax[1].plot(y, 'go-')
ax[1].grid()
ax[1].set_title('$f(x) = -x^3$')
ax[1].axhline(color='blue')
ax[1].axvline(color='blue')

plt.show()
```

```python
print(f'x: {x} \n')
print(f'f(x): {y}')
```

### 3. Creciente

$$
Si \space para \space todo \space x_1, x_2 \in I : x_1 < x_2 \Longleftrightarrow f(x_1) <= f(x_2)
$$

Se lee como:
> Si para todo $x_1$ y $x_2$ que pertenecen al intervalor $I$, tal que $x_1$ sea menor a $x_2$ si y solo si la función de $x_1$ es menor o igual a la función de $x_2$

Siempre aumenta o pertenece constante a su conjunto de definición.

Un ejemplo es $f(x) = e^x$, es creciente en todo su dominio, ya que el valor aumenta exponencialmente a medida que $x$ aumenta.

**Comprobación**

```python
N = 10

def f(x):
    return np.exp(x)

x = np.linspace(1, 5, num=N)
y = f(x)

fig, ax = plt.subplots(1, 2)
ax[0].plot(x, 'ro-')
ax[0].grid()
ax[0].set_title('x')
ax[0].axhline(color='blue')
ax[0].axvline(color='blue')

ax[1].plot(y, 'go-')
ax[1].grid()
ax[1].set_title('$f(x) = e^x$')
ax[1].axhline(color='blue')
ax[1].axvline(color='blue')

plt.show()
```

### 4. Decreciente

$$
Si \space para \space todo \space x_1, x_2 \in I : x_1 < x_2 \Longleftrightarrow f(x_1) >= f(x_2)
$$

Se lee como:
> Si para todo $x_1$ y $x_2$ que pertenecen al intervalor $I$, tal que $x_1$ sea menor a $x_2$ si y solo si la función de $x_1$ es mayor o igual a la función de $x_2$

Siempre disminuye o pertenece constante a su conjunto de definición.

Un ejemplo es $f(x) = log(x)$, es creciente en todo su dominio de números reales positivos, ya que el valor disminuye a medida que $x$ aumenta.

```python
N = 10

def f(x):
    return np.log(x)

x = np.linspace(1, 5, num=N)
y = f(x)

fig, ax = plt.subplots(1, 2)
ax[0].plot(x, 'ro-')
ax[0].grid()
ax[0].set_title('x')
ax[0].axhline(color='blue')
ax[0].axvline(color='blue')

ax[1].plot(y, 'go-')
ax[1].grid()
ax[1].set_title('f(x) = log(x)')
ax[1].axhline(color='blue')
ax[1].axvline(color='blue')

plt.show()
```
