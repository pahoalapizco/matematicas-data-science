```python
import numpy as np
import matplotlib.pyplot as plt
```

# Maniputación de funciones

## Desplazamientos verticales y horizontales

Siendo $c$ una constante mayor que $0$, entonces la gráfica:
- $y = f(x) + c$ se desplaza $c$ unidades hacia arriba.
- $y = f(x) + c$ se desplaza $c$ unidades hacia abajo.
- $y = f(x - c)$ se desplaza $c$ unidades hacia la derecha.
- $y = f(x + c)$ se desplaza $c$ unidades hacia la izquierda.

```python
N = 1000

def f(x):
    return x ** 2

c = 5
x = np.linspace(-5, 5, num=N)

fig, ax = plt.subplots()
ax.plot(x, f(x))
ax.grid()
ax.axhline(color='r')
ax.axvline(color='r')
ax.set_title('f(x)')

plt.show()
```

### Desplazamiento vertical

```python
fig, axes = plt.subplots(1,2)

y_up = f(x) + c
y_down = f(x) - c

# Desp. hacia arriba
axes[0].plot(x, y_up)
axes[0].set_title('f(x) + c, Arriba')
axes[0].grid()
axes[0].axhline(color='r')
axes[0].axvline(color='r')

# Desp. hacia abajo
axes[1].plot(x, y_down)
axes[1].set_title('f(x) - c, Abajo')
axes[1].grid()
axes[1].axhline(color='r')
axes[1].axvline(color='r')


fig.tight_layout()
plt.show()
```

### Desplazamiento horizontal

```python
fig, axes = plt.subplots(1,2)

y_left = f(x + c) # Izquierda
y_righ = f(x - c) # Derecha

# Desp. hacia derecha
axes[0].plot(x, y_righ)
axes[0].set_title('f(x - c), Derecha')
axes[0].grid()
axes[0].axhline(color='r')
axes[0].axvline(color='r')

# Desp. hacia izquierda
axes[1].plot(x, y_left)
axes[1].set_title('f(x + c), Izqerda')
axes[1].grid()
axes[1].axhline(color='r')
axes[1].axvline(color='r')

fig.tight_layout()
plt.show()
```

## Alargamiento y compresiones

Siendo $c$ una constante mayor que $0$, entonces la gráfica:
- $y = f(x) * c$ alarga la gráfica verticalmente en un factor de $c$.
- $y = f(x) * \frac 1 c$ comprime la gráfica verticalmente en un factor de $c$.
- $y = f(x * c)$ comprime la gráfica horizontalmente en un factor de $c$.
- $y = f(x * \frac 1 c)$ alarga la gráfica horizontalmente en un factor de $c$.

```python
N = 1000

def f(x):
    return np.sin(x)

x = np.linspace(-15, 15, num=N)
y = f(x)


fig, ax = plt.subplots()
ax.plot(x, f(x))
ax.grid()
ax.axhline(color='r')
ax.axvline(color='r')
ax.set_title('f(x)')

plt.show()
```

### Alargamiento

```python
fig, axes = plt.subplots(1,2)

c = 3

y_v = f(x) * c # Vertical
y_h = f(x * (1/c)) # Horizontal

# Alarg. Vertical
axes[0].plot(x, y_v)
axes[0].set_title('f(x) * c, Vertical')
axes[0].grid()
axes[0].axhline(color='r')
axes[0].axvline(color='r')

# Alarg. Horizontal
axes[1].plot(x, y_h)
axes[1].set_title('f(x * (1/c)), Horizontal')
axes[1].grid()
axes[1].axhline(color='r')
axes[1].axvline(color='r')

fig.tight_layout()
plt.show()
```

### Compresión

```python
fig, axes = plt.subplots(1,2)

c = 3

y_v = f(x) * (1/c) # Vertical
y_h = f(x * c) # Horizontal

# Compr. Vertical
axes[0].plot(x, y_v)
axes[0].set_title('f(x) * (1/c) c, Vertical')
axes[0].grid()
axes[0].axhline(color='r')
axes[0].axvline(color='r')

# Compr. Horizontal
axes[1].plot(x, y_h)
axes[1].set_title(' f(x * c), Horizontal')
axes[1].grid()
axes[1].axhline(color='r')
axes[1].axvline(color='r')

fig.tight_layout()
plt.show()
```

## Reflexiones

- $y = -f(x)$ refleja la grafica respecto al eje $x$
- $y = f(-x)$ refleja la grafica respecto al eje $y$

```python
N = 100

def f(x):
    return x ** 2

x = np.linspace(-10, 10, num=N)
y = f(x)

fig, ax = plt.subplots()

ax.plot(x, y)
ax.grid()
ax.set_title('f(x)')
ax.axhline(color='red')
ax.axvline(color='red')
plt.show()
```

```python
fig, ax = plt.subplots(1, 2)

y_x = -f(x)
y_y = f(-x)

ax[0].plot(x, y_x)
ax[0].grid()
ax[0].set_title('-f(x), eje x')
ax[0].axhline(color='red')
ax[0].axvline(color='red')

ax[1].plot(x, y_y)
ax[1].grid()
ax[1].set_title('f(-x), eje y')
ax[1].axhline(color='red')
ax[1].axvline(color='red')

plt.show()
fig.tight_layout()
```
