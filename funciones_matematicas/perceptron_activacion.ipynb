# Perceptron

Es un tipo básico de neurona artifical utilizado en redes neuronales artificiales.
<br>


![Perceptron](./imgs/perceptron.png)

El perceptron se compone de:
- **Señales de entrada:** Características.
- **Pesos sinápticos:** Valor ponderado de cada características.
- **Unión sumadora:** Suma de la operación de multiplicar la sseñales de entrada con sus pesos ponderados, también se le llama _combinación líneal_, como salida nos da una linea recta.
- **Función de activación:** Función matemática que se aplica al resultado de la unión sumadora para obtener una salida no lineal y así poder darle mayor complejidad que le permitirá a la neurona separar ó dividir los datos.

> El perfectron funciona tomando las señales de entrada, multiplicandolas por sus respectivos pesos sinápticos y luego sumandolas, a la suma resultante (unión sumadora) se le aplica una función de activación para determinar la salida del perceptron.

## Funciones de activación

Las funciones de activación normalizan el resultado de la (las) unión sumadora con el fin de evitar rangos drasticamente diferentes, la activación en redes neuronales, lo que hace es convertir los valores anteriores a algo equivalente a -1,1 ó 0,1, de esta forma hacen todo el proceso estadísticamente equilibrado.

### Función lineal

$$
f(x) = mx + b
$$

Devuelve la entrada sin realizar ninguna transformación. Se utiliza en casos en los que se requiere una salida linal. <br>
**Ejemplo:** Problemas de regresión lineal

```python
def f(x):
    return x

plt.plot(x, f(x))
plt.grid()
plt.show()
```

### Función escalón de Heaviside
$$H(x) = 
     \begin{cases}
        0, &\quad \text{para, } x < 0 \\
        1,  &\quad\text{para. } x \ge 0 \\
     \end{cases}
$$

Es una función que devuelve 1 si la entrada es mayor o igual a 0, y 0 en caso contrario. Se utiliza en problemas de clasificación binaria.

```python
def H(x):
    y = np.zeros(len(x))
    for idx, val in enumerate(x):
        if val >= 0:
            y[idx] = 1
    return y

plt.plot(x, H(x))
plt.grid()
plt.show()
```

### Función sigmoide

$$
f(x) = \frac {1}{1 + e^{-x}}
$$

También conocida como función logistica. Esta función tiene forma de "_S_" y mapea la entrada en rangos de valores entre $0$ y $1$. Utilizada comunmente en problemas de clasificación binaria.

```python
def s(x):
    return 1 / (1 + np.exp(-x))

plt.plot(x, s(x))
plt.grid()
plt.show()
```

### Función de tangente hiperbólica

$$
f(x) = {\frac {2}{1 + e^{-2x}}} - 1
$$

Esta función tiene una forma similar a la funciómm _sigmoide_, pero mapea la entrada a un rango de -1 y 1. Al igual que la función _sigmoide_, se utiliza en problemas de clasificación.

```python
def th(x):
    return np.tanh(x) # Numpy ya tiene una función que implementa la formula matematica de la tangente hiperboliza

plt.plot(x, th(x))
plt.grid()
plt.show()
```

### Función ReLU
Rectified Linear Unit por sus siglas en ingles
$$
R(x) = max(0, x)
$$

_ReLU_ es una función que devuelve la entrada si es mayor a 0, y 0 en caso contrario.
Es ampliamente utilizada en redes neuronales debido a su simplicidad y efectividad en el procesamiento de los datos. Se utiliza en problemas de clasificación y también en capas ocultas para introducir no linealidad.

```python
def R(x):
    return np.maximum(x, 0)

plt.plot(x, R(x))
plt.grid()
plt.show()
```
