# Funciones trascendentes
Son funciones que no pueden ser expresadas con polinomios.

## Funciones trigonometricas

Agunos ejemplos de este tipo de funciones son:
- $cos(x)$
- $sen(x)$
- $tan(x)$

En ciencia de datos se utilizan para el procesamiento de señales, visualización de datos, interpolación y ajuste de modelos.

### Coseno

```python
def f(x):
    return np.cos(x)

y = f(x)

plt.plot(x, y)
plt.grid()
```

### Seno

```python
def f(x):
    return np.sin(x)

y = f(x)

plt.plot(x, y)
plt.grid()
```

### Tangente

```python
def f(x):
    return np.tan(x)

y = f(x)

plt.plot(x, y)
plt.grid()
```

## Función exponencial

Tiene la forma:
$$f(x) = a^x$$

Donde la base $a$  es una constante positiva. Un gran ejemplo de una función exponencial es usando la base como el número de euler:
$$f(x) = e^x$$

```python
def f(x):
    return np.exp(x)

y = f(x)

plt.plot (x, y)
plt.grid()
```

## Función logaritmo

El logaritmo esta definido por la siguiente **relación**:
$$log_{b}(x) = n \Longleftrightarrow x = b^n $$

donde:
- **$b$** es la base.
- **$n$** es el exponente al que está elevado la base.
- **$x$** es el resultado de elevar la base $b$ al exponente $n$.

**Ejemplo**: <br>
Teniendo $b=2$ y $n=8$, entnces:
$$2^8 = 256$$

Por lo tanto $x = 256$. Calculando el logaritmo base 2 de $x$ tenemos:
$$log_{2}(256) = 8$$

```python
def f(x):
    return np.log2(x) # El calculo de un logaritmo se hace a travez de metodos numéricos, sin embargo numpy ya tiene implementados dichos metodos en sus funciones de logaritmos.
```

Una de las reglas de los logaritmos es que no podemos tener un valor de $x$ menor o igual a 0. <br>
Por lo tanto generaremos otro array de $x$ solo con numeros mayores a $0$.

```python
x = np.linspace(0.1, 256, num=N)
y = f(x)

plt.plot(x, y)
plt.grid()
```

Una de las caracteristicas de los logaritmos esque a mayor volumen de datos, mas tarda en crecer la línea en el grafico.
