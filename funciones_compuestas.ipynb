# Funciones compuestas
Son aquellas que se crean combinando do o m치s funciones.

Se utiliza la salida(rango) de una funci칩n como entrada(dominio) para otra funci칩n. <br>
En terminos matem치ticos se expresa como:

$$ f \space o \space g = (f \space o \space g)(x) = f(g(x))$$

```python
N = 1000

# funciones:
def g(x):
    return x**2

def f(x):
    return np.cos(x)

x = np.linspace(-10, 10, num=N)
f_o_g = f(g(x)) # El resultado de la funci칩n g para como argumento a la funci칩n f

plt.plot(x,f_o_g)
plt.grid()
```

Otra forma de implementar las funciones compuestas:

```python
# Para este caso invert칤 el orden de las funciones
y = f(x)
g_o_f = g(y) # la variable y tiene el resultado de la funci칩n f, dicho resultado se pasa como argumento a la funci칩n g

plt.plot(x,f_o_g)
plt.grid()
```
