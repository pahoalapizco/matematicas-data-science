# Visualizaciones

```python
import pandas as pd
import seaborn as sns
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

## Scatterplot
Gráfico en forma de plano cartesiano donde cada punto representa el valor de dos variables, una en el eje `x` y otra en el `y`. Utilizado para encontrar correlaciones entre variables.

Relación entre el tamaño del sepalo y el petalo

```python
sns.scatterplot(data=iris, x="sepal_length", y="petal_length")
```

Segmentamos por especie

```python
sns.scatterplot(data=iris, x="sepal_length", y="petal_length", hue="species")
```

Las flores iris, de especie setosa son más pequeñas que las otras especies de esta misma flor.

## Join plot
Combina un `scatterplot` con un gráfico de frecuencias que nos an una idea de como estan distribuidos los datos y como se relacionan entre si.

```python
sns.jointplot(data=iris, x="sepal_length", y="petal_length", hue="species")
```

## Boxplot
Resumén de los datos donde podemos ver los `quartiles`, `rango` y `outliers`.

```python
sns.boxplot(data=iris, x="species", y="sepal_length", hue="species")
```

```python
sns.barplot(data=iris, x="species", y="sepal_length", hue="species")
```
